```python
def Fileselector(YearList,UrlYear,ComponentList,UrlComponent):
    '''This function allows you to currently specify up to one 
       year and 5 categories for scraping from.This will be 
       updated to allow selection of multiple years and streamlined'''
    AddYear = True                  # While active scrape will query the
                                    # User for a year input 
    AddComponent = True
    YearInput = ''
    ComponentInput = []
    ComponentSelection = ['1','2','3','4','5']
    
    while AddYear:                                              
        print('select one of the following years:\n ', YearList[0:])
        YearInput = input()
        
        if YearInput in YearList:
            UrlYear.append(YearInput)
            AddYear = False
        elif len(YearInput) > 4:
            print('Incorrect entry try again')
        else:
            print('Year unavailable try again')
    
    while AddComponent:
        print('Select numbers for the following categories:\n',
              '1.', ComponentList[0],'\n',
              '2.', ComponentList[1],'\n',
              '3.', ComponentList[2],'\n',
              '4.', ComponentList[3],'\n',
              '5.', ComponentList[4],'\n',)
        ComponentInput = input('select category from 1-5:' )
        
        if ComponentInput in ComponentSelection:
            UrlComponent.append(ComponentList[int(ComponentInput) - 1])
            AddAnother = input('Add another Category? y/n')
            if AddAnother is 'y':
                pass
            elif AddAnother is 'n':
                AddComponent = False
            else:
                print('an error occurred')
                pass
        else:
            print('Incorrect selection try again')
            pass
        
def Filescraper(IDno,Url):
    """Filescraper will extract all NHANES .XPT files from the specified target url. It achieves this by iterating
       over the javascript table and extracting the file href from each individual row. This continues until no further
       files are found."""
    driver.get(Url)
    while True:
        try:
            IDno += 1
            rowID = str(IDno)
            result = driver.find_element_by_xpath(r"//*[@id='GridView1']/tbody/tr[" +rowID+"]/td[3]/a")
            data.append(result.get_attribute("href"))
        except:
            print(str(IDno - 1) + ' files downloaded')
            break

def Filesaver(SaveDir,SaveComponent):
    for file in data:
        urllib.request.urlretrieve(file, filename = (SaveDir + file[32:]))
        print(file[43:] + " Successfully saved")
```
